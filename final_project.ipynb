```python
import pandas as pd
import seaborn
import matplotlib.pyplot as plt 
file_path = 'Student_performance_data _.csv'
student_data = pd.read_csv(file_path, encoding='latin1')
student_data

```

```python
# Test Code
top10students = student_data.sort_values(by = 'GPA', ascending = False)
top10students.head(20)
```

```python
#Question 1 What is the average GPA of students who receive tutoring versus no tutoring?

#Average GPA of Students with Tutoring
gpa_tutoring = student_data[student_data['Tutoring'] == 1]['GPA'].mean()

#Average GPA of Students with No Tutoring
gpa_notutoring = student_data[student_data['Tutoring'] == 0]['GPA'].mean()

print("Average GPA of students who received tutoring:" , gpa_tutoring)
print("Average GPA of students who did not receive tutoring:" , gpa_notutoring)

```

```python
#Question 1 Visualization - Bar Graph
labels = ['Tutoring', 'No Tutoring']
average_gpas = [gpa_tutoring, gpa_notutoring]

plt.bar(labels, average_gpas, color=['blue', 'orange'])

plt.title('Average GPA by Tutoring Status')
plt.xlabel('Tutoring Status')
plt.ylabel('Average GPA')

plt.show()
```

```python
#Question 2 What is the average GPA of students grouped by weekly study hours?


bins = [0, 5, 10, 15, 20]
labels = ['0-5 hours', '6-10 hours', '11-15 hours', '16-20 hours']


#Average GPA of Students of students grouped by weekly study hours
student_data['StudyTimeGroup'] = pd.cut(student_data['StudyTimeWeekly'], bins=bins, labels=labels, include_lowest=True)
average_gpa_by_study_time = student_data.groupby('StudyTimeGroup')['GPA'].mean().reset_index()

print(average_gpa_by_study_time)
```
