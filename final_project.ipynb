```python
import pandas as pd
import seaborn
import matplotlib.pyplot as plt 
file_path = 'Most Streamed Spotify Songs 2024.csv'
music_data = pd.read_csv(file_path, encoding='latin1')
music_data

# Display the head of the dataset, top 20 rows
#top_20 = music_data.head(20)
#print(top_20)
```

```python
music_data['Spotify Streams'] = music_data['Spotify Streams'].str.replace(',', '').astype(float)
```

```python

spotify_columns = music_data[['Track','Album Name', 'Artist', 'Spotify Streams']]


most_streamed_spotify_songs = spotify_columns.sort_values(by='Spotify Streams', ascending=False)

top_50 = most_streamed_spotify_songs.head(50)
top_50


```
