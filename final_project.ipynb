```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt 
file_path = 'Student_performance_data _.csv'
student_data = pd.read_csv(file_path, encoding='latin1')
student_data

```

```python
# Test Code
top10students = student_data.sort_values(by = 'GPA', ascending = False)
top10students.head(20)
```

```python
#Question 1 What is the average GPA of students who receive tutoring versus no tutoring?

#Average GPA of Students with Tutoring
gpa_tutoring = student_data[student_data['Tutoring'] == 1]['GPA'].mean()

#Average GPA of Students with No Tutoring
gpa_notutoring = student_data[student_data['Tutoring'] == 0]['GPA'].mean()

print("Average GPA of students who received tutoring:" , gpa_tutoring)
print("Average GPA of students who did not receive tutoring:" , gpa_notutoring)

```

```python
#Question 1 Visualization - Bar Graph
labels = ['Tutoring', 'No Tutoring']
average_gpas = [gpa_tutoring, gpa_notutoring]

plt.bar(labels, average_gpas, color=['blue', 'orange'])

plt.title('Average GPA by Tutoring Status')
plt.xlabel('Tutoring Status')
plt.ylabel('Average GPA')

plt.show()
```

```python
#Question 2 What is the average GPA of students grouped by weekly study hours?


bins = [0, 5, 10, 15, 20]
labels = ['0-5 hours', '6-10 hours', '11-15 hours', '16-20 hours']


#Average GPA of Students of students grouped by weekly study hours
student_data['StudyTimeGroup'] = pd.cut(student_data['StudyTimeWeekly'], bins=bins, labels=labels, include_lowest=True)
gpa_by_study_hours = student_data.groupby('StudyTimeGroup')['GPA'].mean().reset_index()

print(gpa_by_study_hours)
```

```python
#Question 2 Visualization
plt.figure(figsize=(8, 5))
plt.bar(gpa_by_study_hours['StudyTimeGroup'], gpa_by_study_hours['GPA'], color='skyblue')
plt.xlabel('Weekly Study Hours Group')
plt.ylabel('Average GPA')
plt.title('Average GPA by Weekly Study Hours')
plt.show()
```

```python
#Question 3 Max number of absences? Is there any correlation between the number absences and GPA?

#Max Number of Absences

max_absences = student_data['Absences'].max()
print("Maximum number of absences:", max_absences)
```

```python
#Question 3 Visualization

plt.figure(figsize=(8, 5))
plt.scatter(student_data['Absences'], student_data['GPA'], color='blue', alpha=0.6)
plt.xlabel('Number of Absences')
plt.ylabel('GPA')
plt.title('Scatterplot of GPA vs. Absences')
plt.grid(True)
plt.show()
```

```python
#Question 4 Top 10 Students Versus Bottom 10

top_10_students = student_data.nlargest(10, 'GPA')[[ 'StudyTimeWeekly', 'GPA']]
bottom_10_students = student_data.nsmallest(10, 'GPA')[['StudyTimeWeekly', 'GPA']]

top_10_students
bottom_10_students

```

```python
#Question 4 Visualization

top_10_students['Group'] = 'Top 10'
bottom_10_students['Group'] = 'Bottom 10'
combined_data = pd.concat([top_10_students, bottom_10_students])

plt.figure(figsize=(12, 6))
sb.scatterplot(data=combined_data, x='StudyTimeWeekly', y='GPA', hue='Group', palette={'Top 10': 'blue', 'Bottom 10': 'red'})
plt.xlabel('Weekly Study Hours')
plt.ylabel('GPA')
plt.title('Top 10 vs Bottom 10 Students: Weekly Study Hours vs GPA')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Question 5 What is the average GPA of students who are involved in sports versus those who are not?

average_gpa_by_sports = student_data.groupby('Sports')['GPA'].mean()

average_gpa_by_sports
```

```python
#Question 5 Visualization

plt.figure(figsize=(8, 6))
average_gpa_by_sports.plot(kind='bar', color=['blue', 'orange'])
plt.xlabel('Sports')
plt.ylabel('Average GPA')
plt.title('Average GPA of Students Sports VS No Sports')
plt.xticks(ticks=[0, 1], labels=['No Sports', 'Sports'], rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
#Question 6: What is the average parental support for the top 20 students and the lower 20 students?
students_top_20 = student_data.sort_values(by='GPA', ascending=False).head(20)
students_bottom_20 = student_data.sort_values(by='GPA', ascending=True).head(20)


# Calculate the average parental support for the top 0 students
avg_parent_support_top_20 = students_top_20['ParentalSupport'].mean()

# Calculate the average parental support for the bottom 20 students
avg_parent_support_bottom_20 = students_bottom_20['ParentalSupport'].mean()

print("Average Parent Support Top 20 GPA: ", avg_parent_support_top_20)
print("Average Parent Support Bottom 20 GPA: ", avg_parent_support_bottom_20)
```

```python
#Question 6 Visualization

categories = ['Top 20 Students', 'Bottom 20 Students']

# Values for the bar graph
values = [avg_parent_support_top_20, avg_parent_support_bottom_20]

# Create the bar graph
plt.bar(categories, values, color=['blue', 'red'])

# Add title and labels
plt.title('Average Parental Support: Top 20 vs Bottom 20 Students')
plt.ylabel('Average Parental Support')

# Show the graph
plt.show()
```

```python
#Question 7 How many students have between 0.0 - 1.0 GPA? 1.0 - 2.0? 2.0 - 3.0? 3.0 - 4.0?

gpa_ranges = [(0.0, 1.0), (1.0, 2.0), (2.0, 3.0), (3.0, 4.0)]

# Count the number of students in each GPA range
gpa_counts = {}
for lower_bound, upper_bound in gpa_ranges:
    count = student_data[(student_data['GPA'] >= lower_bound) & (student_data['GPA'] < upper_bound)].shape[0]
    gpa_counts[f'{lower_bound} - {upper_bound}'] = count

# Display the counts for each range
for range_label, count in gpa_counts.items():
    print(f'Number of students with GPA in range {range_label}: {count}')
```

```python
#Question 7 Visualization

gpa_labels = ['0.0 - 1.0', '1.0 - 2.0', '2.0 - 3.0', '3.0 - 4.0']

plt.bar(gpa_labels, gpa_counts, color='skyblue')


plt.title('Number of Students in Each GPA Range')
plt.xlabel('GPA Ranges')
plt.ylabel('Number of Students')

plt.show()
```
